### 4. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.

```python
# Code here 
s1 = pd.Series([1, 2, 3, 4, 5, 6])
s2 = pd.Series([4, 5, 6, 7, 8, 9])

print("Addition:\n", s1 + s2)
print("\nSubtraction:\n", s1 - s2)
print("\nMultiplication:\n", s1 * s2)
print("\nDivision:\n", s1 / s2)
```

### 5. Write a Pandas program to compare the elements of the two Pandas Series.

```python
# Code here (==, >=, <=, !=)
print("Equal (==):\n", s1 == s2)
print("\nGreater or Equal (>=):\n", s1 >= s2)
print("\nLess or Equal (<=):\n", s1 <= s2)
print("\nNot Equal (!=):\n", s1 != s2)
```
