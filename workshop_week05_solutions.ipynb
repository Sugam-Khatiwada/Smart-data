```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Task 1

### 1. Matplotlib – Bar Chart
Create a bar chart showing the number of students in 5 different classes.

```python
classes = ['A', 'B', 'C', 'D', 'E']
students = [25, 30, 22, 28, 35]

plt.figure(figsize=(7, 4))
plt.bar(classes, students, color='steelblue')
plt.title('Number of Students per Class')
plt.xlabel('Class')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

### 2. Seaborn – Bar Plot
Create the same bar chart in Seaborn and color the bars blue.

```python
plt.figure(figsize=(7, 4))
sns.barplot(x=classes, y=students, color='steelblue')
plt.title('Number of Students per Class (Seaborn)')
plt.xlabel('Class')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

### 3. Matplotlib – Box Plot
Generate a box plot for exam scores and highlight the outliers.

```python
scores = [55, 67, 78, 88, 92, 45, 66, 70, 89, 95]

plt.figure(figsize=(5, 5))
plt.boxplot(scores)
plt.title('Exam Scores – Box Plot')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

## Task 2

### 4. Seaborn – Box Plot
Create a box plot in Seaborn for the same scores and change the color to green.

```python
plt.figure(figsize=(5, 5))
sns.boxplot(y=scores, color='green')
plt.title('Exam Scores – Box Plot (Seaborn)')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

### 5. Matplotlib – Scatter Plot
Plot hours studied vs marks obtained.

```python
hours = [2, 3, 4, 5, 6, 7]
marks = [40, 50, 55, 65, 70, 80]

plt.figure(figsize=(7, 4))
plt.scatter(hours, marks, color='coral', s=80)
plt.title('Hours Studied vs Marks Obtained')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.tight_layout()
plt.show()
```

### 6. Seaborn – Scatter Plot with Regression Line
Create a scatter plot for the same data and add a regression line.

```python
plt.figure(figsize=(7, 4))
sns.regplot(x=hours, y=marks)
plt.title('Hours Studied vs Marks (with Regression Line)')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.tight_layout()
plt.show()
```

## Task 3

### 7. Matplotlib – Pie Chart
Create a pie chart for fruits sold.

```python
fruits = ['Apple', 'Banana', 'Orange', 'Mango']
quantity = [50, 30, 15, 5]

plt.figure(figsize=(6, 6))
plt.pie(quantity, labels=fruits,
         autopct='%1.1f%%')
plt.title('Fruits Sold')
plt.tight_layout()
plt.show()
```

### 8. Seaborn – Categorical Box Plot
Plot salaries by department.

```python
departments = ['HR', 'IT', 'Sales', 'Marketing']
salaries = [40000, 50000, 45000, 60000]

plt.figure(figsize=(7, 4))
ax = sns.boxplot(
    x=departments,
    y=salaries,
    hue=departments,
    palette='Set2'
)
plt.title('Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```

### 9. Seaborn – Count Plot
Create a count plot of favourite colors from survey data.

```python
colors = ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Red', 'Green']

plt.figure(figsize=(6, 4))
sns.countplot(
    x=colors,
    hue=colors,
    palette={'Red': 'red', 'Blue': 'blue', 'Green': 'green'},
    legend=False
)
plt.title('Favourite Colors – Count Plot')
plt.xlabel('Color')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
