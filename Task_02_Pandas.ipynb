#### *v) Team1 vs Team2 Dashbord* 

*Create a function which will take two string(name of two teams) as input. Show win Loss record between them <br>
and player getting most player of the match award in matches between these two teams.* <br>
`team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')`

```python
def team1_vs_team2(team1, team2):
    matches = ipl[
        ((ipl['Team1'] == team1) & (ipl['Team2'] == team2))
        | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))
    ]

    if matches.empty:
        print(f'No matches found between {team1} and {team2}.')
        return

    print(f'Total matches between {team1} and {team2}: {len(matches)}')
    print('\nWin/Loss record:')
    print(matches['WinningTeam'].value_counts(dropna=False))

    print('\nTop Player of the Match:')
    print(matches['Player_of_Match'].value_counts().head(1))
```

```python
team1_vs_team2('Kolkata Knight Riders', 'Chennai Super Kings')
```
